1. A random variable \( X \) has the following probability distribution:

| x    | 0  | 1  | 2  | 3  | 4  | 5  | 6  |
|------|----|----|----|----|----|----|----|
| P(x) | a  | 3a | 2a | 5a | 7a | 4a | 8a |

(i) Determine a 
(ii) Find \( P(X > 3) \)
(iii) \( P(X < 4) \)
(iv) \( P(1 < X < 5) \)

Solution:
 
$\sum_{i=0}^6 P(x)=1$ 

ie. P(0) + P(1) + P(2) + P(3) + P(4) + P(5) + P(6) = 1 \
ie. a + 3a + 2a + 5a + 7a + 4a + 8a = 1 \
ie 30a = 1 \
ie. a = 1/30 

substituting value of a we get,

| x    | 0     |   1  |   2  |   3  |   4  |   5  |   6  |
|------|-------|------|------|------|------|------|------|
| P(x) | a     | 3a   |   2a |   5a |   7a |   4a |   8a |
| P(x) | 1/30  | 1/10 | 2/30 | 5/30 | 7/30 | 4/30 | 8/30 |


P(X > 3) = P(4) + P(5) + P(6) \
= 7a + 4a + 8a
= 19a
= 19/30

P(X<4) = 1 - P(X>3)
= 1-19/30 \
= 11/30


 P(1 < X < 5)
 = P(X=2) + P(X=3) + P(X=4) \
 = 2a + 5a + 7a
 = 14 a
 = 14 / 30
 = 7/15

2. A continuous variable \( x \) has pdf \( f(x)=kx^3, \; 0<x<1 \) then i) find the value of \( k \), ii) find \( a \) such that \( P(x < a) = P(x>a) \)

Solution:

First we find the value of a. From the given pdf we know that x over range (0, 1) is given and = 1.
$$
\begin{aligned}
\int_0^1 f(x)\,dx &= 1 \\
\int_0^1 kx^3\,dx &= 1 \\
k \left[ \frac{x^4}{4} \right]_0^1 &= 1 \\
\frac{k}{4} &= 1 \\
k &= 4
\end{aligned}
$$


To find the value of a, we have, 
$$
\begin{aligned}
\int_0^a f(x) dx &= \int_{a}^1 f(x) dx \\
\int_0^a 4x^3\,dx &= \int_a^1 4x^3\,dx\\
4 \left[ x^4 \right]_0^a &= 4 \left[ x^4 \right]_a^1 \\
\left[ a^4 - 0^4  \right] &= \left[ 1^4 - a^4  \right] \\
a^4 &= 1 - a^4 \\
2a^4 &= 1 \\
a &= \left(\frac{1}{2}\right)^{1/4} \\
a &= 0.8409 
\end{aligned}
$$

3. On tossing two fair coins find the expected number of heads.

Solution:

let X = probability of getting head

Finding sample space,

P(X = 0) = P(TT) = 1/2 * 1/2 = 1/4 \
P(X = 1) = P(HT) + P(TH) = 1/2*1/2 + 1/2*1/2 = 1/4 + 1/4 = 1/2 \
P(X = 2) = P(HH) = 1/2 + 1/2 = 1/4

Now expected number of getting head is,

$E(X) = \sum x⋅P(X=x) $ \
= 0 * P(X = 0) + 1 * P(X = 1) + 2P(X = 2) \
= 0 + 1/2 + 2*1/4 \
= 1/2 + 1/2 \
= 1

A continuous random variable \( x \) has the pdf \( f(x)= i x(1-x), \; 0<x<1 \), find \( E(x) \) and \( E(x^2) \).

Solution:

First finding the value of i,

$$
\begin{aligned}
\int_0^1 f(x)\,dx &= 1 \\
\int_0^1 ix(1-x)\,dx &= 1 \\
i \int_0^1 x-x^2\,dx &= 1 \\
i \left[ \frac{x^2}{2} - \frac{x^3}{3}  \right]_0^1 &= 1 \\
i \left[ (\frac{1^2}{2} - \frac{1^3}{3}) - (\frac{0^2}{2} - \frac{0^3}{3})  \right] &= 1 \\
i \left[ \frac{1}{2} - \frac{1}{3} \right] &= 1 \\
i \frac{3-2}{6} &= 1 \\
i \frac{1}{6} &= 1 \\
\therefore i &= 6
\end{aligned}
$$

Substituting value of i, our pdf becomes f(x)= 6x(1-x)

Now for finding E(x), $E(x) = \int_a^b x f(x) \, d(x)$

= $\int_0^1 x.f(x)\,dx $ \
= $\int_0^1 x.6x(1-x)\,dx $ \
= $6 \int_0^1 x^2 - x^3 \,dx $ \
= $6 \left[ \frac{x^3}{3} - \frac{x^4}{4} \right]_0^1 $ \
= $6 \left[ (\frac{1^3}{3} - \frac{1^4}{4}) - (\frac{0^3}{3} - \frac{0^4}{4} ) \right]_0^1 $ \
= $6 \frac{1}{3} - \frac{1}{4}$ \
= $6 . \frac{1}{12}$ \
= $\frac{1}{2}$ 

$$ \therefore E(x) = \frac{1}{2}$$

Similarly for finding $E(x^2)$, $E(x^2) = \int_a^b x^2 f(x) \, d(x)$

= $\int_0^1 x^2.f(x)\,dx $ \
= $\int_0^1 x^2.6x(1-x)\,dx $ \
= $6 \int_0^1 x^3 - x^4 \,dx $ \
= $6 \left[ \frac{x^4}{4} - \frac{x^5}{5} \right]_0^1 $ \
= $6 \left[ (\frac{1^4}{4} - \frac{1^5}{5}) - (\frac{0^4}{4} - \frac{0^5}{5} ) \right]_0^1 $ \
= $6 \frac{1}{4} - \frac{1}{5}$ \
= $6 . \frac{1}{20}$ \
= $\frac{3}{10}$ 

$$ \therefore E(x^2) = \frac{3}{10}$$
