**1. Assume that the sex ratio is 1:1. What is the probability that a family with five children has** \
**(i) no girl** \
**(ii) at least one boy.**

Solution,

Here, \
sex ration = 1:1 \
Probability of being a girl = 1/2 \
Probability of being a boy = 1/2

For a family with five children; n=5, sample space or possible sequences is 2^5 =  = 32

Probability that there is no girl = Probability that there is all boys.
P(BBBBB) = 1/32

Pobability that there is at least on boy is,
1 - Probability that there is all girls = 1 - 1/32 = 31/32 

Alternatively solving using the probability distribution

Here, \
sex ratio = 1:1 \
P(Girl) = 1/2 \
P(Boy) = 1/2

Number of children ie, the no of trials (n) = 5

Let X denotes the no of girls.
then, p = 1/2 and q = 1 - p = 1/2


This is binomial random variable denoted by: \
$ X  \sim \text{Binomial}(n=5, p=1/2) $ \
with probability distribution frunction: \
$ P(x) = \binom{n}{x} p^x q^{n-x}$

a. The probability of getting no girls is \
ie, $ P(X=0) =  \binom{5}{0} p^0 q^{5-0}$  \
ie, $ P(X=0) =  1 * \frac{1}{2}^{5}$ \
ie, $ P(X=0) =  1 * \frac{1}{32}$ 

b. The probability of getting at least one boy is 1 - Probability of getting all girls \
$ = 1 - P(X=5) $ \
$ = 1 -  \binom{5}{5} p^5 q^{5-5}$ \
$ = 1 -  1 * \frac{1}{2}^5 \frac{1}{2}^{0}$ \
$ = 1 -  1 * \frac{1}{32} * 1$ \
$ = \frac{31}{32} $

**2. The probability that an applicant for a driver’s license will pass the license test in any try is 0.75.** \
**i) what is the probability that the applicant will finally pass the license test on the fourth try?** \
**ii) Find the expected number of failed tests before the applicant passes the license test first time.**

Solution,

probability of getting driver's lisence (p) = 0.75 \
q = 1-p = 0.25

Since this is Geometric Distribution, \
PMF = $P(X=x) = q^{x}p$
where x is the number of failures

The probability that applicant will pass on fourth try is; \
3 failures and 1 success
$P(X=3) = q^{3}p$ \
$P(X=3) = 0.25^{3}*0.75$ \
$P(X=3) = 0.0117$


Expected no of failures before success is given by p/q \
$ = \frac{0.75}{0.25}$ \
$ = 0.33$

**5. Assuming that one in 80 births is a case of twins, calculate the probability of 2 or more sets of twins on a day when 30 births occur. Compare the results obtained by using Binomial and Poisson distribution separately.**

Solution,

Given,
probability of twins in a birth = 1/80 = 0.0125
no of births in a day (n) = 30
let X be the no of twins 

We need to find P(X>=2) \
P(X>=2) = 1 - P(X<2) = 1 - [P(X=0) + P(X=1)]

**Assuming the binomial distribution:** 

$ X \sim \text{Binomial}(n=30,p=0.0125) $

P(X=0) \
 = $(1 - q)^n$ \
 = $(1 - 0.0125)^{30} $  \
 = 0.6857 

P(X=1) \
= $\binom{n}{x} p^x (1-p)^{n-x} $ \
= $\binom{30}{1} 0.0125^1 (1-0.0125)^{30-1} $ \
= 0.2605

Therefore, P(X>=2) = 1 - (0.6857 + 0.2605) = 0.0538

**Assuming poisson distribution**

average no of set of twins per 30 births (λ) = np = 30 * 0.0125 = 0.375

$ X \sim \text{Poisson}(λ=0.375) $

P(X=0) \
= $ \frac{\lambda^{x} e^{-\lambda}}{x!}$ \
= $ \frac{0.375^{0} e^{-0.375}}{0!}$ \
= $ e^{-0.375} $ \
= 0.6873

P(X=1) \
= $ \frac{0.375^{1} e^{-0.375}}{1!}$ \
= $ 0.375 * e^{-0.375}$ \
= 0.2577

Therefore, P(X>=2) = 1 - (0.6873 + 0.2577) = 0.055

Both distributon resulted similar outputs, but computation of binomial was tough due to large power (30). Hence we can conclude that if n is large the the binomial distribution can be approximated by poisson distribution.

**6. An automatic machine makes paper clips from coils of wire. On the average, 1 in 400 paper clips is defective. If the paper clips are packed in boxes of 100, assuming that the process follows Poisson distribution, what is the probability that any given box of clips will contain,** \
**(i) no defective,** \
**(ii) one or more defectives,** \
**(iii) less than two defectives?**

Solution,

average defects per 400 clips = 1
average defects per box (100 clips)= 100/400 = 1/4 = 0.25

let x be the no of defects per box 

$ X \sim \text{Poisson}(λ=0.25)$ 

Now, 

Probability that there is no defect P(X=0) = $e^{-\lambda}$ = e^{-0.25} = 0.7788

Probability that there is one or more defect P(X>=1) = 1 - P(X<1) = 1 - P(X=0) = 1 - 0.778 = 0.2212

Probability that there is less than two defect P(X<2) = P(X=0) + P(X=1) \
we calculate P(X=1) first which is $\lambda e^{-\lambda}$ = $0.025 e^{-0.025}$ = 0.25 * 0.7788 = 0.1947 \
Therefore,   P(X<2) = 0.7788 + 0.1947 = 0.9735
